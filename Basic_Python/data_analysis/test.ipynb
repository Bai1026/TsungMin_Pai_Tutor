## è³‡æ–™åˆ†æçš„é¡å‹

**æè¿°æ€§åˆ†æ (Descriptive Analytics)ï¼š**
- **ç›®æ¨™**ï¼šäº†è§£ã€Œç™¼ç”Ÿäº†ä»€éº¼ã€
- **æ–¹æ³•**ï¼šçµ±è¨ˆæ‘˜è¦ã€åœ–è¡¨ã€å ±è¡¨
- **ç¯„ä¾‹**ï¼šæœˆéŠ·å”®å ±å‘Šã€ç¶²ç«™æµé‡çµ±è¨ˆ

**è¨ºæ–·æ€§åˆ†æ (Diagnostic Analytics)ï¼š**
- **ç›®æ¨™**ï¼šäº†è§£ã€Œç‚ºä»€éº¼ç™¼ç”Ÿã€
- **æ–¹æ³•**ï¼šç›¸é—œæ€§åˆ†æã€å› æœåˆ†æ
- **ç¯„ä¾‹**ï¼šç‚ºä»€éº¼æŸç”¢å“éŠ·é‡ä¸‹é™

**é æ¸¬æ€§åˆ†æ (Predictive Analytics)ï¼š**
- **ç›®æ¨™**ï¼šé æ¸¬ã€Œå°‡æœƒç™¼ç”Ÿä»€éº¼ã€
- **æ–¹æ³•**ï¼šæ©Ÿå™¨å­¸ç¿’ã€æ™‚é–“åºåˆ—åˆ†æ
- **ç¯„ä¾‹**ï¼šé æ¸¬ä¸‹å€‹æœˆçš„éŠ·å”®é¡

**è¦ç¯„æ€§åˆ†æ (Prescriptive Analytics)ï¼š**
- **ç›®æ¨™**ï¼šå»ºè­°ã€Œæ‡‰è©²åšä»€éº¼ã€
- **æ–¹æ³•**ï¼šæœ€ä½³åŒ–æ¼”ç®—æ³•ã€æ¨¡æ“¬
- **ç¯„ä¾‹**ï¼šæœ€ä½³å®šåƒ¹ç­–ç•¥ã€åº«å­˜ç®¡ç†

````python
# æè¿°æ€§åˆ†æç¯„ä¾‹
import pandas as pd
import numpy as np

# å»ºç«‹ç¯„ä¾‹è³‡æ–™
np.random.seed(42)
student_scores = {
    'æ•¸å­¸': np.random.normal(75, 15, 100),
    'è‹±æ–‡': np.random.normal(80, 12, 100),
    'ç‰©ç†': np.random.normal(70, 18, 100)
}

scores_df = pd.DataFrame(student_scores)

# æè¿°æ€§çµ±è¨ˆ
print("å­¸ç”Ÿæˆç¸¾æè¿°æ€§çµ±è¨ˆï¼š")
print(scores_df.describe())

# ç›¸é—œæ€§åˆ†æ
print("\nç§‘ç›®é–“ç›¸é—œæ€§ï¼š")
print(scores_df.corr())
````

````python
# 2. è³‡æ–™æ¸…ç†ç¯„ä¾‹
import pandas as pd
import numpy as np

# å»ºç«‹åŒ…å«å•é¡Œçš„ç¯„ä¾‹è³‡æ–™
dirty_data = pd.DataFrame({
    'name': ['Alice', 'Bob', None, 'Diana', 'Eve'],
    'age': [25, -5, 30, 200, 28],  # åŒ…å«ä¸åˆç†å€¼
    'salary': [50000, 60000, None, 75000, 'unknown'],  # æ··åˆé¡å‹
    'city': ['å°åŒ—', 'TAIPEI', 'å°åŒ— ', 'é«˜é›„', 'å°ä¸­']  # æ ¼å¼ä¸ä¸€è‡´
})

print("åŸå§‹è³‡æ–™ï¼ˆåŒ…å«å•é¡Œï¼‰ï¼š")
print(dirty_data)
print(f"\nè³‡æ–™å½¢ç‹€ï¼š{dirty_data.shape}")
print(f"ç¼ºå¤±å€¼çµ±è¨ˆï¼š\n{dirty_data.isnull().sum()}")

# è³‡æ–™æ¸…ç†æ­¥é©Ÿ
def clean_data(df):
    # 1. è™•ç†ç¼ºå¤±å€¼
    df = df.dropna(subset=['name'])  # åˆªé™¤å§“åç‚ºç©ºçš„è¡Œ
    
    # 2. è™•ç†ç•°å¸¸å€¼
    df = df[(df['age'] >= 0) & (df['age'] <= 120)]  # åˆç†å¹´é½¡ç¯„åœ
    
    # 3. è™•ç†è³‡æ–™é¡å‹
    df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
    
    # 4. æ¨™æº–åŒ–æ ¼å¼
    df['city'] = df['city'].str.strip().str.upper()
    
    return df

cleaned_data = clean_data(dirty_data.copy())
print("\næ¸…ç†å¾Œçš„è³‡æ–™ï¼š")
print(cleaned_data)
````

````python
# Pandas ç¯„ä¾‹ï¼šè³‡æ–™æ“ä½œå·¥å…·
import pandas as pd

print("=== Pandas ç¯„ä¾‹ ===")

# å»ºç«‹ DataFrame
data = {
    'product': ['A', 'B', 'C', 'A', 'B'],
    'sales': [100, 150, 200, 120, 180],
    'region': ['North', 'South', 'North', 'South', 'North']
}

df = pd.DataFrame(data)
print("åŸå§‹è³‡æ–™ï¼š")
print(df)

# åˆ†çµ„èšåˆ
grouped = df.groupby('product')['sales'].sum()
print(f"\nå„ç”¢å“ç¸½éŠ·å”®ï¼š\n{grouped}")

# ç¯©é¸è³‡æ–™
high_sales = df[df['sales'] > 120]
print(f"\né«˜éŠ·å”®è¨˜éŒ„ï¼š\n{high_sales}")
````

````python
# Matplotlib ç¯„ä¾‹ï¼šåŸºç¤è¦–è¦ºåŒ–
import matplotlib.pyplot as plt
import numpy as np

print("=== Matplotlib ç¯„ä¾‹ ===")

# å»ºç«‹ç°¡å–®åœ–è¡¨
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.title('æ­£å¼¦å‡½æ•¸åœ–', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('sin(x)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("åœ–è¡¨å·²å»ºç«‹ä¸¦é¡¯ç¤º")
````
